```python
def solution (A, B):
    #0 up, 1 down
    count=0
    survive=[]
    
    for fish in range (len(A)):
        if B[fish]==0:
            while len(survive)!=0 and survive[-1]<A[fish]:
                survive.pop()
        
        if len(survive)==0 and B[fish]==0:
            count+=1
        elif B[fish]==1:
            survive.append(A[fish])
        print (A[fish], B[fish], survive, count )
            
    count+=len(survive)
    
    return count
```

```python
import random
n=1000
A=list(range(n))
random.shuffle(A)
B=[random.randint(0,1) for _ in range (n)]
```

```python
solution (A,B)
```
